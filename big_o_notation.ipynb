- Label your x axis "Elements"
- Label your y axis "Operations"

- Use LaTex notation where possible

```python
import matplotlib.pyplot as plt
import math
```

```python
print(plt.style.available)
```

## Curves to graph
- $y = 0n + 1$ as O(1)
- $y = log(n)$ as O(log n)
- $y = n$ as O(n)
- $y = n * log(n)$ as O(n log n)
- $y = n^2$ as On^2
- $y = 2n$ as O2^n
- $y = n!$ as O(n!)
- $y = n^n$ as O(n^n)

```python
x = [.25 * x for x in range(100)]
x2 = [.25 *x for x in range(1,101)]
x3 = [x for x in range(100)]

Oaxis = 0 

yellowline = [2 * n for n in x]
O1 = [0*n + 1 for n in x]
Ologn = [math.log(n) for n in x2]
On = [n for n in x]
OnLogn = [n * math.log(n) for n in x2]
On_2 = [n**2 for n in x]
O2_n = [2 * n for n in x]
Onfact = [math.factorial(n) for n in x3]
On_n = [n^n for n in x3]

plt.style.use('seaborn-poster')

plt.figure(figsize = (15,13))
plt.plot(x, O1, label = '$O(1)$', c = 'darkgreen', linewidth = 5)
plt.plot(x, Ologn, label = '$O(log n)$', c = 'green', linewidth = 5)
plt.plot(x, On, label = '$O(n)$', c = 'yellowgreen', linewidth = 5)
plt.plot(x, OnLogn, label = '$O(n log n)$', c = 'orange', linewidth = 5)
plt.plot(x, On_2, label = '$On^2$', c = 'red', linewidth = 5)
plt.plot(x, O2_n, label = '$O2^n$', c = 'yellow', linewidth = 5)
plt.plot(x, Onfact, label = '$O(n!)$', c = 'darkred', linewidth = 5)
plt.plot(x, On_n, label = '$O(n^n)$')

plt.fill_between(x, Ologn, Oaxis,
                 color='green',       # The outline color
                 alpha=0.4)          # Transparency of the fill
plt.fill_between(x, O2_n, Ologn,
                 edgecolor = 'none',
                facecolor = 'yellow',
                 alpha=0.2)
plt.fill_between(x, On_2, O2_n,
                facecolor = 'orange',
                alpha=0.2)
plt.fill_between(x, On_2, 100,     #FINALLY, this above the line thing took forever! 
                facecolor = 'red',
                alpha=0.2)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.xlabel('Elements', fontname = 'Arial Black', fontsize = 25)
plt.ylabel('Operations', fontname = 'Arial Black', fontsize = 25)
plt.title('Big O Notation', fontname = 'Arial Black', fontsize = 40)
plt.xlim(0, 25)
plt.ylim([0, 100])

plt.show()
```
