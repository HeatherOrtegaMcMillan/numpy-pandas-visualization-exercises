# Matplotlib Exercises
1. Use matplotlib to plot the following equation:

$ y = x^2 - x + 2 $

- You'll need to write the code that generates the x and y points.

- Add an anotation for the point 0, 0, the origin.

```python
import matplotlib.pyplot as plt
import math
```

```python
# create range for x
x = list(range(-25, 25))

# set y = to list of values in relation to x using list comprehension
# this basically takes every value in x and plugs it into my formula and puts it in the list
y = [(n ** 2 - n + 2) for n in x]

# add annotation for the origin
#       vvvv this changes location of actual annotation
plt.text(-1, 150, '(0, 0)', fontsize=10, color='red')
plt.annotate('Origin', xy = (0,0), xytext = (0, 100), arrowprops = {})

#plot the thing! I did it! 
plt.plot(x, y)

```

Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

- $y = \sqrt{x}$
- $y = x^3 $
- $y = 2^x$
- $y = 1/(x + 1) $

You can use functions from the math module to help implement some of the equations above.

```python
x1 = list(range(50))
y1 = [math.sqrt(n) for n in x1]
plt.title('$y = \sqrt{x}$')


plt.plot(x1, y1)
```

```python

y2 = [n ** 3 for n in x]
plt.title('$y = x^3$')

plt.plot(x,y2)
```

```python

y3 = [2 ** n for n in x]
plt.title('$y = 2^x$')

plt.plot(x,y3)
```

```python

y4 = [1/(n + 1) for n in x1]
plt.title('$y = 1/(x+1)$')

plt.plot(x,y4)
```

```python
plt.figure(figsize = (15, 10))
plt.subplot(221)
plt.title('$y = \sqrt{x}$')
plt.plot(x, y1)

plt.subplot(222)
plt.title('$y = x^3$')
plt.plot(x, y2)

plt.subplot(223)
plt.title('$y = 2^x$')
plt.plot (x, y3)

plt.subplot(224)
plt.title('$y = 1/(x+1)$')
plt.plot (x, y4)

plt.suptitle('All The Graphs')
plt.show()
```

4. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize = (12,8))
plt.plot(x, y1, c = 'red', label = '$\sqrt{x}$')
plt.plot(x, y2, c = 'purple', label = '$x^3$')
plt.plot(x, y3, c = 'green', label = '$2^x$')
plt.plot(x, y4, c = 'blue', label = '$1/(x+1)$')

plt.legend(loc='upper right')
plt.title('All the Graphs in One')


plt.xlim([0,10])
plt.ylim([0,150])
plt.show()
```
